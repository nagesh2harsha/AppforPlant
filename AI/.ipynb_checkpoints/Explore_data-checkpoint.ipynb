```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from shutil import copyfile
```

```python
path=r"C:\work\freelancing\plants-collage-project\PlantVillage-Dataset\raw\color"
```

```python
img=[]
for i in os.listdir(path):
    #print(os.path.join(os.path.join(path,i),os.listdir(os.path.join(path,i))[0]))
    img.append(plt.imread(os.path.join(os.path.join(path,i),os.listdir(os.path.join(path,i))[0])))
print(len(img))
```

```python
fig,ax=plt.subplots(7,6,figsize=(20,15))
c=0
for i in range(7):
    for j in range(6):
        if c==len(os.listdir(path)):
            break
        ax[i][j].imshow(img[c])
        ax[i][j].set_title(os.listdir(path)[c])
        c+=1
plt.show()
```

```python
plt.imshow(img[0])
```

```python
d=dict()
c=0
labels=os.listdir(path)
for i in labels:
    d[i]=np.array(img[c]).flatten()
    c+=1
len(d)
```
